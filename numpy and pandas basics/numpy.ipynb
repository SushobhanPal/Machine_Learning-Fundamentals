```python
import numpy as np
arr=np.array([1,2,3,4,5,6])
add=arr+5
print(add)
mul=arr*5
print(mul)
exp=arr**2
print(exp)

```

```python
import numpy as np
arr=np.array([1,2,3,4,5,6])
print("element at index 2: ",arr[2])
print("array slicing from index 1 to 4",arr[1:5])
print("array slicing using negative index",arr[-1:-5:-1])
```

```python
import numpy as np
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[2,3,4],[7,6,3]])
mul=arr1*arr2
mul
```

```python
import numpy as np
arr=np.array([1,2,3,4,5,6])
print(np.min(arr))
print(np.argmin(arr))
print(np.max(arr))
print(np.argmax(arr))
```

```python
import numpy as np
arr1=np.array([[1,2,3],[4,5,6]])
arr2=np.array([[2,3,4],[7,6,3]])
con=np.concatenate((arr1,arr2),axis=0)
print(con)
stk=np.vstack((arr1,arr2))
stk2=np.hstack((arr1,arr2))
print(stk)
print(stk2)
```
